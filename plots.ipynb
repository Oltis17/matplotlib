# Knižnica matplotlib

## Potrebné includy

````python
%matplotlib inline  
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
````

## Prvý jednoduchý graf

````python
x = [0, 1, 2, 3]
y = [0, 1, 2, 3]

plt.plot(x,y)

plt.show()
````

## Sinusoida

````python
x = np.arange(0, 4 * np.pi, 0.1)   # start, stop, step
y = np.sin(x)

plt.plot(x, y, color="red")
plt.figure(figsize=(8, 8), dpi=80)

plt.show()
````

## Stĺpcový graf

````python
data = [[30, 25, 50, 20],
[40, 23, 51, 17],
[35, 22, 45, 19]]
X = np.arange(4)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.bar(X + 0.00, data[0], color = 'b', width = 0.25, label = "blue")
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(X + 0.5, data[2], color = 'r', width = 0.25)
ax.set_title("Názov grafu")
ax.set_xlabel('Scores')
ax.set_ylabel('Scores')

ax.set_xticks(np.arange(4))

fig.legend()
````

## Kombinácia dvoch grafov

````python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'

sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax = plt.subplots(2)
ax[0].pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

data = [[30, 25, 50, 20],
[40, 23, 51, 17],
[35, 22, 45, 19]]
X = np.arange(4)


ax[1].bar(X + 0.00, data[0], color = 'b', width = 0.25, label = "blue")
ax[1].bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax[1].bar(X + 0.5, data[2], color = 'r', width = 0.25)
ax[1].set_title("Názov grafu")
ax[1].set_xlabel('Scores')
ax[1].set_ylabel('Scores')

ax[1].set_xticks(np.arange(4))

fig.legend()

plt.show()
````

import matplotlib
import matplotlib.pyplot as plt


def normalize(arr):
    new_x = []

    for i in arr:
        if i == arr[-1]:
            i = i[:-1]
        new_x.append(int(i))
    return new_x

f = open("text.txt", "r")

x = []
y = []

count = 0

for line in f:
    if count == 0:
        x = line.split(" ")
    else:
        y = line.split(" ")

    count += 1

x = normalize(x)
y = normalize(y)

f.close()

plt.grid(color='gray', linestyle='-', linewidth=0.5)
plt.plot(x, y)
plt.savefig("obr.png")
